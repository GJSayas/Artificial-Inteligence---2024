```python
#1. Loading your friend's data into a dictionary 

# Create the years and durations lists
data = {
    'year': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'duration': [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
    }

# Create a dictionary with the two lists
movie_dict = {'Release Year': data['year'],'Duration (minutes)': data['duration']}

# Print the dictionary
movie_dict
```

```python
#2. Creating a DataFrame from a Dictionary 

#Import pandas under its usual alias

import pandas as pd

#Create a Data Frame from the dictionary
durations_df=pd.DataFrame(movie_dict)

#Print the DataFrame
durations_df
```

```python
#3. A visual inspection of our data

# Import matplotlib.pyplot under its usual alias and create a figure

import matplotlib.pyplot as plt

fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(durations_df['Release Year'], durations_df['Duration (minutes)'], color='purple')

# Create a title
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.title('How have Movie Durations changed Over the Years')


# Show the plot
plt.show()
```

```python
#4. Loading the rest of the data from a CSV

# Read in the CSV as a DataFrame
netflix_df = pd.read_csv('DATASETS/TAREA 2/netflix_data.csv')

# Print the first five rows of the DataFrame
netflix_df.head(5)
```

```python
#5. Filtering for movies

# Subset the DataFrame for type "Movie"
netflix_df_movies = netflix_df[netflix_df['type']=='Movie']

# Select only the columns of interest
netflix_movies_col_subset = netflix_df_movies[['title', 'country','genre','release_year','duration']]

# Print the first five rows of the new DataFrame
netflix_movies_col_subset.head(10) 
```

```python
#6. Creating a scatter plot

# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year

duration = netflix_movies_col_subset['duration']
release_year = netflix_movies_col_subset['release_year']

plt.scatter(release_year, duration, alpha=0.5, color='purple')

# Create a title

plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.title('How have Movie Durations changed Over the Years')

# Show the plot
plt.show()
```

```python
#7. Digging deeper

# Filter for durations shorter than 60 minutes

duration_less_60min = netflix_movies_col_subset['duration'] < 60
short_movies = netflix_movies_col_subset[duration_less_60min]


# Print the first 20 rows of short_movies
short_movies.head(20)
```

```python
#8. Marking non-feature films


# Define an empty list
colors = []

# Iterate over rows of netflix_movies_col_subset and make a color for each genre

for genre in netflix_movies_col_subset['genre']:
    if genre == 'Children':
        colors.append('yellow')
        
    elif genre == 'Stand-Up':
            colors.append('orange')
            
    elif genre == 'Documentaries':
            colors.append('cyan')
    else:
            colors.append('gray')

# Inspect the first 10 values in your list  

colors[:10]
```

```python
#9. Plotting with color!

# Set the figure style and initalize a new figure
plt.style.use('seaborn-v0_8-pastel')
fig =  plt.figure(figsize=(12, 8))

# Create a scatter plot of duration versus release_year
plt.scatter(release_year, duration, alpha=0.5, color=colors)
# Create a title and axis labels

plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.title('How have Movie Durations changed Over the Years')

# Show the plot
plt.show()
```
